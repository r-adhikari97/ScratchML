# ScratchML : K Nearest Neighbour 
- Developing KNN Model using logic
- #### STEPS :
    1. For a Given Data point
    2. Calculate distance from all other data points in dataset
    3. Get the closest K points
    4. Based on Model -
        - **Regression** : Extract Average of thier values
        - **Classification** : Extract the label with Majority votes

## Model Buildup

```python
import numpy as np
from collections import Counter
```

```python
# Setting up class

def euclidan_distance(x1, x2):
    return np.sqrt(np.sum((x1-x2)**2))


class KNN:
    def __init__(self, k=3): # Setting Default value to 3
        self.k = k

    
    def fit(self, X, y):
        """
        Storing values for X and y datasets
        """
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        predictions = [self._predict(x) for x in X]
        return predictions

    def _predict(self,x):
        # Computing Distances : Euclidian Distance
        distances = [euclidan_distance(x,x_train) for x_train in X_train]

        # Extract Closest K
        k_indices = np.argsort(distances)[:self.k]
        k_nearest_labels = [self.y_train[i] for i in k_indices]

        # Determine Output with majority vote
        most_common = Counter(k_nearest_labels).most_common()
        return most_common[0][0]

    def acc_score(self, predictions, y_test):
        return np.sum((predictions == y_test)/len(y_test))
```

## Setting Up Model

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
cmap = ListedColormap(['#FF0000','#00FF00','#0000FF'])
```

```python
iris = load_iris()
X,y = iris.data, iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1234)
```

```python
# Plotting data
plt.figure(figsize=(12,8))
plt.scatter(X[:,2], X[:,3], c=y, cmap=cmap, s=20, edgecolor='k')
plt.show()
```

## KNN Classifier

```python
clf = KNN(k=5)
clf.fit(X_train, y_train)
preds = clf.predict(X_test)
```

```python
# Computing accuracy
print(f"Accuracy Score : {clf.acc_score(preds, y_test):.3f}")
```
